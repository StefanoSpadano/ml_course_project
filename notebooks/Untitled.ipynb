# Intro + a linear regression model
In this notebook I will attempt at replicating a simple linear regression model

```python
#import libraries used to build a linear regression model

import numpy as np
import matplotlib.pyplot as plt
```

I will be using house pricings from the city where I was born, Pescara in Italy. I found these data online and I will use them as a training dataset to develop the linear regression model. 

```python
#x_training is the size of the house expressed in square meters; here acts as the input variable
#y_training is the price of the house expressed in euros 
#and is written as the product between the square meters size of the house and the average price for that neighborood of the city;
#here it acts as the output variable
x_training = np.array([45, 75, 100, 120, 150, 200])
y_training = np.array([45*2100, 75*2100, 100*2563, 120*1615, 150*2100, 200*2100])
```

```python
#print the two training data just defined
print(f"x_training data is: {x_training}")
print(f"y_training data is: {y_training}")
```

```python
#save m as the number of training examples given
print(f"x_training shape is {x_training.shape}")
m = x_training.shape[0]
print(f"The number of training examples given is {m}")
```

```python
#show all the training data just declared
for i in range(m):
    x_i = x_training[i]
    y_i = y_training[i]
    print(f"The {i}-th data in the training sample is: (x^({i}), y^({i})) = ({x_i}, {y_i})")
```

## Scatter plot

I'm going now to plot these data in a scatter plot

```python
#Plot the data points
plt.scatter(x_training, y_training, marker = 'x', c='g') #choose x as marker and green as color
#set a title
plt.title("Housing prices of Pescara (different zones")
#set axes labels
plt.ylabel("Prices in euros")
plt.xlabel("Size in square meters")
plt.show()
```

## Model function

The model function for linear regression could be expressed in this way: $f_{wb}(x^{i}) = wx^{i} + b$ and changing the values of w and b will give us different lines. Now we are going to define a function that computes the model output once we pass it as parameters the size in square meters alongside the parameters w and b. 



```python
def compute_model_output(x, w, b):
    """
    This function computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```

Now we first define the parameters w and b and then call the model and try to plot it.

```python
#these are two first guesses we can make (I started from w = 100 and b = 100 as it was online), but they can always be changed after we plot the model and see if it fits with the data we have
w = 2500
b = 500
print(f"w: {w}")
print(f"b: {b}")

tmp_f_wb = compute_model_output(x_training, w, b,)

# Plot our model prediction
plt.plot(x_training, tmp_f_wb, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_training, y_training, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in euros)')
# Set the x-axis label
plt.xlabel('Size (m^2)')
plt.legend()
plt.show()
```

Now that we have something similar to a proper model we can use this values to make a prediction for the housing price of a house with some determined size in square meters; let's we are interested in a house that has 137 m^2 size. 

```python
x_i = 137
cost_137_m2_house = w * x_i + b

print(f"For a house measuring {x_i} square meters in Pescara, considering an average price per square meters (averaged over different zones of the city) the price could be something like {cost_137_m2_house} euros.")
```
